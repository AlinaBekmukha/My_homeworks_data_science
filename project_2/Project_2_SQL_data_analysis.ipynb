# Unit 4. Performing Detailed Analysis

1. Write a query that will allow you to find out how many (cnt) vacancies in each region (area) are. Sort by the number of vacancies in descending order.

```python
# query
query_4_1 = f'''select
            a.name title,
            count(v.area_id)
            FROM public.vacancies v
            JOIN public.areas a ON v.area_id = a.id
            
            GROUP BY title
            ORDER BY count DESC
            LIMIT 5
        '''
        
df_4_1 = pd.read_sql_query(query_4_1, connection)
df_4_1
```

### 2. Write a query to determine how many vacancies have at least one of the two salary fields filled.

```python
# query
query_4_2 = f'''select
            COUNT(id)
            FROM public.vacancies
            WHERE salary_from IS NOT NULL OR salary_to IS NOT NULL
        '''
        
df_4_2 = pd.read_sql_query(query_4_2, connection)
df_4_2
```

### 3. Find the average values ​​for the lower and upper limits of the salary ranges. Round the values ​​to the nearest integer.

```python
# query
query_4_3 = f'''select
            ROUND(AVG(salary_from)) lowest,
            ROUND(AVG(salary_to)) highest
            FROM public.vacancies
        '''
        
df_4_3 = pd.read_sql_query(query_4_3, connection)
df_4_3
```

### 4. Write a query that will display the number of vacancies for each type of work schedule and employment type used in vacancies. Sort the result in descending order.

```python
# query
query_4_4 = f'''select
            schedule,
            employment,
            COUNT(employment) jobs_quantity
            FROM public.vacancies
            
            GROUP BY schedule, employment
            ORDER BY COUNT(employment) DESC
        '''
        
df_4_4 = pd.read_sql_query(query_4_4, connection)
df_4_4
```

### 1. Write a query that will allow you to find out which employers are in first and fifth place in terms of the number of vacancies.

```python
# query
query_5_1 = f'''select
            e.name,
            COUNT(v.id)
            FROM public.employers e
            JOIN public.vacancies v ON v.employer_id = e.id
            
            GROUP BY 1
            ORDER BY 2 DESC
            LIMIT 5
        '''
        
df_5_1 = pd.read_sql_query(query_5_1, connection)
df_5_1
```

### 2. Write a query that for each region will display the number of employers and vacancies in it. Among the regions in which there are no vacancies, find the one with the largest number of employers.

```python
# query
query_5_2 = f'''select
            a.name area_name,
            COUNT(e.id) cnt_employers,
            COUNT(v.id) cnt_vacancies
            FROM public.areas a
            LEFT JOIN public.vacancies v ON a.id=v.area_id
            LEFT JOIN public.employers e on a.id = e.area
            WHERE v.id IS NULL
            
            GROUP BY area_name
            ORDER BY cnt_employers DESC
        '''
        
df_5_2 = pd.read_sql_query(query_5_2, connection)
df_5_2
```

### 3. For each employer, count the number of regions in which they publish their vacancies. Sort out the result in descending order.

```python
# query
query_5_3 = f'''select
            e.name employer_name,
            COUNT(DISTINCT v.area_id) cnt_areas
            FROM public.employers e
            JOIN public.vacancies v ON e.id=v.employer_id
            
            GROUP BY 1
            ORDER BY 2 DESC
        '''
        
df_5_3 = pd.read_sql_query(query_5_3, connection)
df_5_3
```

### 4. Write a query to count the number of employers who do not have a 'field of activity' filled.

### 5. Write a query to find out the name of the company that is in third place in the alphabetical list (by name) of companies that have four areas of activity.

```python
# query
query_5_5 = f'''select
            e.name employers_name,
            COUNT(ei.industry_id) cnt_industries
            FROM public.employers e 
            LEFT JOIN public.employers_industries ei ON e.id = ei.employer_id
            
            GROUP BY employers_name
            HAVING COUNT(ei.industry_id) = 4
            ORDER BY employers_name ASC
            LIMIT 3
            
        '''
        
df_5_5 = pd.read_sql_query(query_5_5, connection)
df_5_5
```

### 7. For the Yandex company, display a list of million+ population regions in which the company's vacancies are presented, along with the number of vacancies in these regions. Also add the line with the total number of vacancies in the company. Sort the result in ascending order.

```python
!pip install requests
!pip install beautifulsoup4

import requests 
from bs4 import BeautifulSoup

url = 'https://ru.wikipedia.org/wiki/%D0%93%D0%BE%D1%80%D0%BE%D0%B4%D0%B0-%D0%BC%D0%B8%D0%BB%D0%BB%D0%B8%D0%BE%D0%BD%D0%B5%D1%80%D1%8B_%D0%A0%D0%BE%D1%81%D1%81%D0%B8%D0%B8' 
response_wiki = requests.get(url) 
page = BeautifulSoup(response_wiki.text, 'html.parser') 
wiki_a = page.find_all('a') 
cities = [city.text for city in wiki_a[90:106]]

cities_tuple = tuple(cities)
cities_tuple
```

```python
# query
query_5_7 = f'''SELECT
                a.name cities, 
                count(v.id) cnt_vacancies
                from 
                public.vacancies v 
                left join public.areas a on a.id = v.area_id 
                left join public.employers e on v.employer_id = e.id 
                where e.name = 'Яндекс' 
                and a.name in {cities_tuple} 
                
                GROUP BY cities
                
                union all
                
                select 
                'Total', 
                count(v.id) 
                from 
                public.vacancies v 
                left join public.areas a on a.id = v.area_id 
                left join public.employers e on v.employer_id = e.id 
                where e.name = 'Яндекс' 
                and a.name in {cities_tuple}
                ORDER BY cnt_vacancies
                
        '''
        
df_5_7 = pd.read_sql_query(query_5_7, connection)
df_5_7
```

### 6.4 Check how popular Python is in DS employer job postings requirements. To do this, calculate the number of job openings that list Python as a key skill.

### 6.5 How many key skills are listed on average in job postings for DS? Round your answer to two decimal digits after the point.

```python
# query
query_6_5 = f'''select
            ROUND(AVG(LENGTH(key_skills) - LENGTH(REPLACE(key_skills, CHR(9), '')) + 1), 2) average
            FROM public.vacancies
            
            WHERE (lower(name) LIKE '%data scientist%'
                OR lower(name) LIKE '%data science%' 
                OR lower(name) LIKE '%исследователь данных%' 
                OR (name LIKE '%ML%' AND name NOT LIKE '%HTML%')
                OR lower(name) LIKE '%machine learning%'
                OR lower(name) LIKE '%машинн%обучен%')
                
                AND (key_skills IS NOT NULL)
        '''
        
df_6_5 = pd.read_sql_query(query_6_5, connection)
df_6_5
```

### 6.6 Write a query to calculate the average salary for DS for each type of required experience (a unique value from the Experience field).

```python
# query
query_6_6 = f'''select
            experience,
            ROUND(AVG(COALESCE((salary_to + salary_from)/2, salary_from, salary_to))) average
            FROM public.vacancies
            
            WHERE (lower(name) LIKE '%data scientist%'
                OR lower(name) LIKE '%data science%' 
                OR lower(name) LIKE '%исследователь данных%' 
                OR (name LIKE '%ML%' AND name NOT LIKE '%HTML%')
                OR lower(name) LIKE '%machine learning%'
                OR lower(name) LIKE '%машинн%обучен%')
                
                AND (salary_from IS NOT NULL OR salary_to IS NOT NULL)
                AND (experience IS NOT NULL)
            
            GROUP BY experience
        '''
        
df_6_6 = pd.read_sql_query(query_6_6, connection)
df_6_6
```
